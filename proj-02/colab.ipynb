```python
dirmodel = './'
modelhdr =  open(dirmodel+'gaussian_anomaly.hdr','r')
# Read the first line of header file
[nz,ny,oz,oy,dz,dy] = modelhdr.readline().strip().split()
nz, ny = int(nz), int(ny)  # Convert n1 and n2 to integers
oz,oy,dz,dy = float(oz), float(oy), float(dz), float(dy)
# Read the second line of header file
velbin = modelhdr.readline().strip()
print(velbin)
f =  open(dirmodel+velbin,'rb')
print(f)
modelnpy = np.fromfile(f,dtype='f4',count=nz*ny)
modelnpy = modelnpy.reshape([nz, ny],order='F')
f.close()

# Construct z and y axes arrays
z_values = np.arange(oz, oz + nz * dz, dz)
y_values = np.arange(oy, oy + ny * dy, dy)

# Obtain max and min values
maxval = np.max(modelnpy)
minval = np.min(modelnpy)

# Plot the image
plt.imshow(modelnpy, cmap='jet', extent=[y_values[0], y_values[-1], z_values[-1], z_values[0]])
# Adjust color scale limitsi
plt.clim(vmin=minval, vmax=maxval)
# Add colorbar
cbar = plt.colorbar(shrink=0.5)
cbar.ax.tick_params(labelsize=10)
tick_values = np.arange(minval, maxval + 1, 250)
cbar.set_ticks(tick_values)

# Add title and labels
plt.title('Gaussian anomaly model')
plt.xlabel('Distance (m)')
plt.ylabel('Depth (m)')
# Set colorbar label on top
cbar.set_label(r'$V_p$ (m/s)', labelpad=-18, y=1.16, rotation=0)

plt.show()
```
